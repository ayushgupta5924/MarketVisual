````python
import yfinance as yf

# Search for the company
results = yf.Search("HDFC Bank").quotes

for result in results:
    # FILTER: Only process if the exchange is NSE (NSI) or BSE
    if result['exchange'] in ['NSI', 'BSE']:
        
        symbol = result['symbol']
        ticker = yf.Ticker(symbol)
        
        # Use try-except to handle cases where price data isn't immediately available
        try:
            current_price = round(ticker.fast_info['last_price'], 2)
            print(f"Symbol: {symbol:<15} | Price: {current_price:<10} | Name: {result['shortname']}")
        except:
            print(f"Symbol: {symbol:<15} | Price: N/A        | Name: {result['shortname']}")
````

````python
import yfinance as yf

stock_name = input("Enter the stock name (e.g., Tata, HDFC, Adani): ")

results = yf.Search(stock_name).quotes

print("-" * 85)
print(f"{'Symbol':<20} | {'Price':<10} | {'Exchange':<10} | {'Name'}")
print("-" * 85)

for result in results:
    if result['exchange'] in ['NSI', 'BSE']:
        
        symbol = result['symbol']
        exchange = result['exchange']
        name = result['shortname']
        
        try:
            ticker = yf.Ticker(symbol)
            price = round(ticker.fast_info['last_price'], 2)
            
            print(f"{symbol:<20} | {price:<10} | {exchange:<10} | {name}")
            
        except:
            print(f"{symbol:<20} | {'N/A':<10} | {exchange:<10} | {name}")
````

````python
endDate = dt.datetime.now()

startDate = endDate - dt.timedelta(days = 365)
endDate, startDate
````

````python
stocks = ['RELIANCE.NS', 'NIFTYBEES.NS', 'HDFCBANK.NS', 'TCS.NS']
````

````python
df = yf.download(stocks, start = startDate, end = endDate)
print(df.tail())
````

````python
# Using the `yfinance` library, we download the historical stock data for our selected stocks within the specified date range.

# **Exploring the Dataframe**

# ```python
df.head()
````

````python
adj_close_prices = df['Close']
print("\n=== Daily Close Prices for Selected Stocks ===")
adj_close_prices.head()
````

````python
adj_open_prices = df['Open']
print("\n=== Daily Open Prices for Selected Stocks ===")
adj_open_prices.head()
````

````python
adj_high_prices = df['High']
print("\n=== Daily High Prices for Selected Stocks ===")
adj_high_prices.head()
````

````python
adj_low_prices = df['Low']
print("\n=== Daily Low Prices for Selected Stocks ===")
adj_low_prices.head()
````

````python
# To calculate the daily returns, we can use the pct_change() function from pandas
log_returns = np.log(adj_close_prices/adj_close_prices.shift(2))
log_returns
````
